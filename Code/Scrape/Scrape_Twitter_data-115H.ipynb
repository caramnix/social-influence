```python
#takes in tweet and outputs dict for ONE tweet. 
def get_dict_for_tweet(tweet_of_interest, tweet_dict, author, Housedat, leg_dict, filename):
    EMPTY ={}
    try: 
        n_mentions= len(tweet_of_interest["entities"]["mentions"])
        #print(n_mentions)
    except KeyError: 
        print(filename)
        return(tweet_dict)
    if len(tweet_of_interest["entities"]["mentions"]) == 0: 
        return (EMPTY)
    else: 
        mentions_dict= {}
        count=0
        for i in range (0, n_mentions): 
            user_mentioned = tweet_of_interest["entities"]["mentions"][i]["username"]
            if user_mentioned in Housedat["screen_name"].tolist(): #if user mentioned is a house member add them to list in dict. #QUESTION, WILL old @'s be used/ignored? 
                try: 
                    leg_dict[author].append(user_mentioned)
                    mentions_dict[count] = user_mentioned
                    count= count+1
                except KeyError: 
                    print("author: " + str(author))
                    print("user_mentioned: " + str(user_mentioned))
                    print("filename" + str(filename))
                    pass
        if (len(mentions_dict) ==0): 
            return(EMPTY)
            
                    
    return({"author": author, "mentions": mentions_dict})
```
