```python
#116 House Twitter
os.chdir("/Users/caranix/Library/CloudStorage/OneDrive-TheOhioStateUniversity/Social Influence/Data/Twitter/116/House")
f = open('Twitter_dict_116th_house_legd.json')
H_116_legd = json.load(f)
```

```python
master_data= pd.read_csv("/Users/caranix/OneDrive - The Ohio State University/2020-2021/Spring/7560/other/merged_data_master.csv")
#u_id_nums= np.unique(master_data["ID"])
```

```python
def get_id_from_handle(handle):
    try:
        return(master_data.loc[master_data["handle"] == handle]["ID"].to_numpy()[0])
    except IndexError: 
        if handle == "repdonyoung": 
            return(1256)
        if handle == "RepArmstrongND": 
            return(2417)
        if handle == "PeterWelch": 
            return(1879)
        if handle == "RepDustyJohnson": 
            return(2439)
        if handle == "RepLizCheney": 
            return(2356)
        if handle == "RepLBR": 
            return(2313)
        if handle =="hornforcongress": 
            return(2431)
        if handle =="DennyHeck":
            return(2170)
        if handle =="JoeKennedy":
            return(2172)
        if handle =="SeanDuffyWI": 
            return(2072)
        if handle =="tomreedny23":
            return(1982)
        if handle =="DonnaShalala":
            return(2391)
        if handle =="RepJamesComer":
            return(2297)
        if handle =="benmcadams": 
            return(2452)
        
    
```

```python
#FOWLER VERSION OF COSPONSORSHIP 
#attach cosponsors to sponsors in directed fashion 

DICT= H_116_legd
u_id_nums= np.unique(master_data["ID"])
probs = np.array([2313, 1256, 2417,1879,2439, 2356])
u_id_nums = np.append(u_id_nums,probs)


adjf_ = pd.DataFrame(index=u_id_nums, columns= u_id_nums)
adjf_ = adjf_.fillna(0)

for key in DICT.keys(): 
    author_id= get_id_from_handle(key)
    n_mentions= len(DICT[key])
    for i in range(0,n_mentions): 
        mention_id = get_id_from_handle(DICT[key][i])
        #print(author_id,mention_id)
        adjf_.loc[int(author_id)][int(mention_id)] += 1
        
```

```python
adjf_.to_csv("adj_mat_Twitter_House_116.csv")
```
