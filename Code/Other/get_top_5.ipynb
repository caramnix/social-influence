```python
#function which returns top 5 connections of inputed congressperson
def pull_top_five(dijkstra_mat,ID):
    top_5_for_ID= dijkstra_mat.loc[int(ID)].sort_values(ascending=True)[0:6].to_frame() #pulls ROW of leg id, sorts so SMALLEST first
    for i in range(0,len(top_5_for_ID)): 
        ID_oi= top_5_for_ID.iloc[i].name
        print(name_from_id(ID_oi))
               
```

```python
def make_data1(Bill_Dict):
    #make helpful dict w/ congressperson info
    data1= {}
    for key in Bill_Dict: 
        sponser_id= Bill_Dict[str(key)]['sponser']['ID']   
    
        if sponser_id not in data1.keys(): 
            data1[int(sponser_id)]= [Bill_Dict[str(key)]['sponser']['LAST NAME'].upper(), Bill_Dict[str(key)]['sponser']['PARTY'], Bill_Dict[str(key)]['sponser']['FULL NAME'], sponser_id]
        num_cospo= len(Bill_Dict[str(key)]['cosponsers'])
        for i in range(0,num_cospo): 
            cospo_id= Bill_Dict[str(key)]['cosponsers'][str(i)]["ID"]
            #print("cospo ID" +str(cospo_id))
            if cospo_id not in data1.keys():
                data1[int(cospo_id)]= [Bill_Dict[str(key)]['cosponsers'][str(i)]['LAST NAME'].upper(), Bill_Dict[str(key)]['cosponsers'][str(i)]['PARTY'],Bill_Dict[str(key)]['cosponsers'][str(i)]['FULL NAME'], cospo_id]
    return (data1)
```

```python
#gets top 5 most connected! 
def get_top_5(connectednesscsv, Bill_Dict,string_name): 
    connectednesscsv= connectednesscsv.sort_values(by=['Connectedness'], ascending= False)
    d1= make_data1(Bill_Dict)
    print(string_name)
    for i in range(0,5): 
        print(data1[int(connectednesscsv.iloc[i].ID)],connectednesscsv.iloc[i]["Connectedness"])
    
```
